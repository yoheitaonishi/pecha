```python
import time
from PIL import Image
import pytesseract
```

## 検出結果

```python
pytesseract
```

```python
image_path = '../images/POODL_Flyer_Design_ja_190322_back_rgb_300dpi.tif'
image = Image.open(image_path)
image
```

```python
print(pytesseract.image_to_osd(image_path))
```

```python
start = time.time()
boxes = pytesseract.image_to_boxes(image_path)
print(boxes)
detect_time = time.time() - start
print('detect_time: ', detect_time)
```

```python
start = time.time()
data = pytesseract.image_to_data(image_path)
print(data)
detect_time = time.time() - start
print('detect_time: ', detect_time)
```
